#Dictionnaire de données - Ore Quality Prediction

##Paramètres mesurés

|Paramètres | Unité | Plage Observée | Role|
|-----------|-------|-----------------|-----|
|Mn         |%      | 30 - 43%        |Critère principal de qualité |
|Fe         |%      | 5 - 20%        |Critère secondaire |

## Règle de classification

- **Haute qualité** : Mn >= 43%
- **Moyenne qualité** : Mn <= 38% et Mn < 43%
- **Basse qualité** : Mn < 38%

Note : le Fe est mesuré et conservé dans la dataset mais n'intervient pas dans la règle de classification - le Mn est le critère déterminant.
